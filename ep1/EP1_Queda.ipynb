# Gráficos dos Dados de Experimento e Simulado

```python
#funcoes para facilitar a mostra dos dados
def easyPlot2D(x,y,title):
  plt.figure()
  plt.plot(x,y)
  plt.title(title)
  plt.show()
```

```python
#funcoes dos vetores de estado
def initStateVector(s):
  return(np.array(s))

def initSVTrajectory():
  return([])

def updateSVTrajectory(svt,s):
  svt.append(s)
  return(svt)

def extractSVTrajectory(svt,i):
  foo = np.array(svt)
  return(foo[:,i])

#Formulas
def Fd(v):
  return K*v**2/M

#Experimental
def updateStateVectorExperimental(s,id):
  dt = sobreposto[0][id+1] - sobreposto[0][id]
  x_e = s[0] + s[1] * dt
  v_e = s[1] + (sobreposto[1][id]-1)*G * dt
  a_e = (sobreposto[1][id]-1)*G
  t_e = s[3] + dt
  return(np.array([x_e,v_e,a_e,t_e]))

#Simulacao com Euler
def updateStateVectorEuler(s,dt):
  return(s + rates(s,dt))

def rates(s,dt):
  r0 = s[1] * dt #movimento no eixo y
  r1 = s[2] * dt #mudança de velocidade no eixo y
  r2 = (-G + Fd(s[1])) - s[2] #mudança de acerelação
  r3 = dt #mudança de tempo
  return(np.array([r0,r1,r2,r3]))
```

```python
#Simulação
def main():
  #Dados Experimentais
  x_e = 0
  v_e = 0
  a_e = 0
  t_e = 0
  stateVectorExperimental = initStateVector([x_e,v_e,a_e,t_e])
  svtExperimental = initSVTrajectory() 

  for i in range(0,399):
    svtvExperimental = updateSVTrajectory(svtExperimental,list(stateVectorExperimental))
    stateVectorExperimental = updateStateVectorExperimental(stateVectorExperimental,i)

  vx_e = extractSVTrajectory(svtExperimental,0)
  vv_e = extractSVTrajectory(svtExperimental,1)
  va_e = extractSVTrajectory(svtExperimental,2)
  vt_e = extractSVTrajectory(svtExperimental,3)

  easyPlot2D(vt_e, vx_e, 'Experimental - tempo, altura')
  easyPlot2D(vt_e, vv_e, 'Experimental - tempo, velocidade')
  easyPlot2D(vt_e, va_e, 'Experimental - tempo, acerelação')

  #--------------------------------------------------------------------------------------------------------
  printSeparador()
  #--------------------------------------------------------------------------------------------------------

  #Tempo da Simulacao
  t_s = 0
  tf = Mtempo
  dt = 0.001

  #Dados Simulados com Euler
  x_s = 0
  v_s = 0
  a_s = -G
  stateVectorEuler = initStateVector([x_s,v_s,a_s,t_s])
  svtEuler = initSVTrajectory() 
  
  while (stateVectorEuler[3]<tf):    
    svtvEuler = updateSVTrajectory(svtEuler,list(stateVectorEuler))
    stateVectorEuler = updateStateVectorEuler(stateVectorEuler,dt)

  vx_s = extractSVTrajectory(svtEuler,0)
  vv_s = extractSVTrajectory(svtEuler,1)
  va_s = extractSVTrajectory(svtEuler,2)
  vt_s = extractSVTrajectory(svtEuler,3)

  easyPlot2D(vt_s, vx_s, 'Simulado - tempo, altura')
  easyPlot2D(vt_s, vv_s, 'Simulado - tempo, velocidade')
  easyPlot2D(vt_s, va_s, 'Simulado - tempo, acerelação')

  #--------------------------------------------------------------------------------------------------------
  printSeparador()
  #--------------------------------------------------------------------------------------------------------
  
  #Comparação
  plt.figure()
  plt.plot(vt_s, vx_s)
  plt.plot(vt_e, vx_e)
  plt.title("Gráfico de comparação da Posição")
  plt.show()

  plt.figure()
  plt.plot(vt_s, vv_s)
  plt.plot(vt_e, vv_e)
  plt.title("Gráfico de comparação da Velocidade")
  plt.show()

  plt.figure()
  plt.plot(vt_s, va_s)
  plt.plot(vt_e, va_e)
  plt.title("Gráfico de comparação da Aceleração")
  plt.show()

main() 
```
