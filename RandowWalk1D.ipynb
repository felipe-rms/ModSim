## Passeios aleatórios (Random Walks) em 1D 
### Teoria e aplicações

### Passeio aleatório em 1D

Imagine um processo iterativo que segue a fórmula:

$y_n = y_{n-1} \pm 1, n>1$

Assim, dado $y_1$, se somar, ou subtrair 1 fosse regido por um processo aleatório, teríamos a sequência:

$y_1, y_2, y_3, \ldots, y_N$ 

que é conhecido como passeio do bêbado, ou passeio aleatório.

Simular um passeio aleatório é muito simples:

```python
import random as rng
import matplotlib.pyplot as plt

def randowWalk(y=0,N=10000):
    Y = []
    Y.append(y)
    for i in range(1,N):
        Y.append(Y[i-1]+rng.choice([-1,1]))
    return Y

plt.plot(randowWalk())
plt.ylabel("Posição do bêbado",fontsize=16)
plt.xlabel('tempo',fontsize=16)
plt.title('Passeio do bêbado',fontsize=16)
```

### Modelagem do passeio aleatório em 1D

Se chamarmos de $d$ a distância do ponto de início do passeio após $N$ passos, isto é, $d=y_1+y_2+\ldots+y_N$, onde estará o bêbado?

Para modelarmos esse processo, vamos ter que usar novamente a ideia de conjuntos estatísticos, ou "ensembles". Assim, ao invés de modelar $d$ vamos modelar $<d>$. Para facilitar o entendimento e as contas, vamos assumir que $y_1=0$. 

$<d> = <y_1+y_2+\ldots+y_N> = \sum\limits_{i=1}^N <y_i>$

Onde $y_i = \pm 1, i>1$, com a probabilidade de somar 1, ou subtrair 1 igual a $\frac{1}{2}$.  

Quando $N$ tende a infinito, como as probabilidade de somar 1, ou -1 são as mesmas, $<y_i> = y_1 = 0$. Ou seja, no limite, o bêbado não sai do lugar. 

E se calcularmos a média do quadrado da distância, isto é, o quanto, na média o bêbado se distancia do ponto inicial em relação à $N$? 

$<d^2> = <(y_1+y_2+\ldots+y_N)^2> = \sum\limits_{i=1}^N <y_i^2> + 2\sum\limits_{i=1}^N\sum\limits_{j=2}^{N-1} <y_iy_j>$

Vamos primeiro calcular $<y_i^2> = (-1)(-1)$ ou $<y_i^2> = 1.1$, portanto  $<y_i^2> = 1$

Depois, por outro lado, temos $y_iy_j$, que pode ser igual a 1, se $y_i=1, y_j=1$, ou $y_i=-1, y_j=-1$ e pode ser igual a -1 se $y_i=1, y_j=-1$, ou $y_i=-1, y_j=1$. Ou seja, $<y_iy_j>=0$.

Portanto, $<d^2> = \sum\limits_{i=1}^N 1 = N$. 

Finalmente, respondendo a pergunta, na média, o bêbado se distancia de $\sqrt{<d^2>} = \sqrt N$
